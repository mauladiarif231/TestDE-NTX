```python
# Import libraries
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder

# Load dataset from CSV file
dataset_path = 'ecommerce-session-bigquery.csv'
df = pd.read_csv(dataset_path)

# Convert 'date' column to datetime format
df['time'] = pd.to_datetime(df['time'], unit='s')

# Scenario 1: Identify top products based on the total transaction revenue per day
daily_revenue = df.groupby(['time', 'v2ProductName'])['totalTransactionRevenue'].sum().reset_index()
top_products_per_day = daily_revenue.groupby('time').apply(lambda x: x.nlargest(100, 'totalTransactionRevenue')).reset_index(drop=True)

print("Top products based on total transaction revenue per day:")

# Set the option to display all columns
pd.set_option('display.max_columns', None)

# Display the DataFrame as a table
display(top_products_per_day.head(50))
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.impute import SimpleImputer

# Mengisi nilai NaN dengan median
imputer = SimpleImputer(strategy='median')
df_filled = imputer.fit_transform(df[['transactions']])

# Inisialisasi model Isolation Forest
clf = IsolationForest(contamination=0.05)

# Fit dan prediksi anomali berdasarkan jumlah transaksi
df['anomaly_score'] = clf.fit_predict(df_filled)

# Analisis hasil anomali
anomalies = df[df['anomaly_score'] == -1]
print("Detected anomalies:")
print(anomalies)

anomalies.to_csv('test2.csv', index=False)
```
